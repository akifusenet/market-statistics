```
# Import pandas 
import pandas as pd   

# reading csv file     
url = 'https://raw.githubusercontent.com/akifusenet/market-statistics/master/EURUSD_m30_last_1_months_UTC.csv'
df = pd.read_csv(url, header=[0]) 
df
```

```
df.loc[:,'DateTime'] = pd.to_datetime(df.Date.astype(str)+' '+df.Timestamp.astype(str))

#Set datetime as index and drop datetime column.
df=df.set_index('DateTime',drop=True)
#Remove redundant columns
df=df.drop(['Date','Timestamp'],axis=1)
df
```
